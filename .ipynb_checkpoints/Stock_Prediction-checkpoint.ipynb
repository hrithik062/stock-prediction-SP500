```python
sp500=sp500.rename(columns={'Open':'Open_sp','High':'High_sp','Low':'Low_sp','Close':'Close_sp','Volume':'Volume_sp','Tomorrow':'Tomorrow_sp','Target':'Target_sp'})
```

```python
sp500.index=sp500.index.date
```

```python
sp500_1=sp500.copy()
sp500_1
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)
train = sp500.iloc[:-100]
test=sp500.iloc[-100:]

predictors = ["Close_sp", "Volume_sp", "Open_sp", "Low_sp", "High_sp"]
model.fit(train[predictors], train["Target_sp"])
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score

preds = model.predict(test[predictors])
```

```python


preds = pd.Series(preds, index=test.index)
```

```python
preds
```

```python
precision_score(test["Target_sp"],preds)
```

```python
combined = pd.concat([test["Target_sp"], preds],axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors],train["Target_sp"])
    preds= model.predict(test[predictors])
    preds=pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target_sp"],preds],axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test= data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target_sp"], predictions["Predictions"])
```

```python
predictions["Target_sp"].value_counts()/predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]

new_predictors =[]
for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    ratio_column=f"Close_Ratio_{horizon}"
    sp500[ratio_column]=sp500["Close_sp"]/rolling_averages["Close_sp"]
    trend_column = f"Trend_{horizon}"
    sp500[trend_column]=sp500.shift(1).rolling(horizon).sum()["Target_sp"]
    new_predictors+=[ratio_column, trend_column]
```

```python
new_predictors
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors],train["Target_sp"])
    preds= model.predict_proba(test[predictors])[:,1]
    preds[preds>=.6] = 1
    preds[preds<.6] = 0
    preds=pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target_sp"],preds],axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target_sp"],predictions["Predictions"])
```

## Finding the correlation between S&P500 and BSE Sensex Index (India)

```python
bse = yf.Ticker("^BSESN")
bse = bse.history(period="max",interval="1d")
bse
```

```python
bse.index
```

```python
bse.plot.line(y="Close", use_index=True)
```

```python
del bse["Dividends"]
del bse["Stock Splits"]
```

```python
bse["Tomorrow"] = bse["Close"].shift(-1)
```

```python
bse
```

```python
bse["Target"]=(bse["Tomorrow"]>bse["Close"]).astype(int)
```

```python
bse
```

```python
bse=bse.rename(columns={'Open':'Open_bse','High':'High_bse','Low':'Low_bse','Close':'Close_bse','Volume':'Volume_bse','Tomorrow':'Tomorrow_bse','Target':'Target_bse'})
```

```python
bse=bse.loc["2005-01-01":].copy()
```

```python
bse
```

```python
bse.index=bse.index.date
```

```python
bse
```

```python
correlation_bse = sp500_1["Target_sp"].corr(bse["Target_bse"])
correlation_bse
```

```python
sp_sg_bse=sgx.join(bse, how='inner').join(sp500_1, how='inner')
```

```python
sp_sg_bse
```

```python
horizons = [2,5,60,250,1000]

predictors_2 =[]
for horizon in horizons:
    rolling_averages = sp_sg_bse.rolling(horizon).mean()
    ratio_column_sp=f"Close_Ratio_sp_{horizon}"
    ratio_column_sg=f"Close_Ratio_sg_{horizon}"
    ratio_column_bse=f"Close_Ratio_bse_{horizon}"
    sp_sg_bse[ratio_column_sp]=sp_sg_bse["Close_sp"]/rolling_averages["Close_sp"]
    sp_sg_bse[ratio_column_sg]=sp_sg_bse["Close_sg"]/rolling_averages["Close_sg"]
    sp_sg_bse[ratio_column_bse]=sp_sg_bse["Close_bse"]/rolling_averages["Close_bse"]
    trend_column_sp = f"Trend_sp_{horizon}"
    trend_column_sg = f"Trend_sg_{horizon}"
    trend_column_bse = f"Trend_bse_{horizon}"
    sp_sg_bse[trend_column_sp]=sp_sg_bse.shift(1).rolling(horizon).sum()["Target_sp"]
    sp_sg_bse[trend_column_sg]=sp_sg_bse.shift(1).rolling(horizon).sum()["Target_sg"]
    sp_sg_bse[trend_column_bse]=sp_sg_bse.shift(1).rolling(horizon).sum()["Target_bse"]
    predictors_2+=[ratio_column_sp,ratio_column_sg,ratio_column_bse, trend_column_sp,trend_column_sg,trend_column_bse]
```

```python
sp_sg_bse=sp_sg_bse.dropna()
```

```python
sp_sg_bse.info()
```

```python
sp_sg_bse
```

```python
predictions = backtest(sp_sg_bse, model, predictors_2)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target_sp"],predictions["Predictions"])
```
